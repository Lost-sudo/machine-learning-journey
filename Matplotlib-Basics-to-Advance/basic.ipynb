```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, linestyle="--", marker="o")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple Line Plot")
plt.grid(True)
plt.show()
```

```python
y2 = [5, 15, 20, 25]
plt.plot(x, y, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5), dpi=100)
plt.plot(x, y)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [10, 20, 30]

plt.bar(labels, values)
plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.scatter(x, y)
plt.title("Plot 2")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x, y)
ax.set_title("OO Style Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

axs[0].plot(x, y)
axs[0].set_title("Line")

axs[1].scatter(x, y)
axs[1].set_title("Scatter")

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(1, 100, 100)
y = np.exp(x)

plt.plot(x, y)
plt.yscale("log")
plt.show()
```

```python
plt.plot(x, y)
plt.annotate("Peak", xy=(3, 25), xytext=(2, 28),
             arrowprops=dict(arrowstyle="->"))
plt.show()
```

```python
epochs = range(1, 11)
loss = [0.9, 0.7, 0.55, 0.42, 0.35, 0.3, 0.27, 0.25, 0.23, 0.22]

plt.plot(epochs, loss, marker='o')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.grid(True)
plt.show()
```
