Copy vs View (COMMON BUG)

```python
a = np.array([1, 2, 3])
b = a           # reference
c = a.copy()    # deep copy
```

```python
b[0] = 100
print(b)
```

```python
c[0] = 999
print(a)
```

Linear Algebra (ML Core)

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
```

```python
A @ B   # Matrix Multiplication
```

```python
np.dot(A, B)
```

```python
np.linalg.det(A)
```

```python
np.linalg.inv(A)
```

```python
np.linalg.eig(A)
```

Statistics and Data Prep

```python
data = np.array([10, 20, 30, 40])
```

```python
# Normalization
(data - data.mean()) / data.std()
```

```python
# Min-Max Scaling
(data - data.min()) / (data.max() - data.min())
```
